```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing dataset
fuel_econ = pd.read_csv('fuel-econ.csv')
```

## Scatterplots with Jitter

```python
#resizing to accomodate two plots
plt.figure(figsize=[20,5])

#plot on left - simple scatter plot
plt.subplot(1,2,1)
sns.regplot(data=fuel_econ, x='year', y='comb')
```

```python
#resizing using truncate = False in regplot to include all points

#resizing to accomodate two plots
plt.figure(figsize=[20,5])

#plot on left - simple scatter plot
plt.subplot(1,2,1)
sns.regplot(data=fuel_econ, x='year', y='comb', truncate=False);

#plot on the right - scatterplot with jitter (only available on sns.regplot)
plt.subplot(1,2,2)
sns.regplot(data=fuel_econ, x='year', y='comb', truncate=False, x_jitter=0.3);
```

## adding transparency to further investigate density of points

```python
#using regplot
sns.regplot(data=fuel_econ, x='year', y='comb', truncate=False, x_jitter=0.3, 
            scatter_kws={'alpha':1/15});
```

```python
#using plt.scatter
plt.scatter(data=fuel_econ, x='year', y='comb', alpha=1/15);
```

```python
#plotting over each other
#using regplot
sns.regplot(data=fuel_econ, x='year', y='comb', truncate=False, x_jitter=0.3, 
            scatter_kws={'alpha':1/20});
#using plt.scatter
plt.scatter(data=fuel_econ, x='year', y='comb', alpha=1/20);
```
