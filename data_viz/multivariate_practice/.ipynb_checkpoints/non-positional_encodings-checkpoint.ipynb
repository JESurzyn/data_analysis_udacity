```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Encoding via Shape

```python
fuel_econ = pd.read_csv('fuel-econ.csv')
```

```python
fuel_econ.info()
```

```python
#cleaning data
fuel_econ.trans[0].split()[0]
```

```python
fuel_econ['trans_type'] = fuel_econ['trans'].apply(lambda x: x.split()[0])
fuel_econ.trans_type.value_counts()
```

```python
#getting a sample
fuel_econ_sample = fuel_econ.sample(200)
fuel_econ_sample.info()
```

```python
#encoding via shape
#generally it is better to assign markers via list of lists rather than dictionary as order is built in
'''
Unfortunately, there is no built-in way to automatically assign different shapes in a single call of the 
scatter or regplot function. Instead, we need to write a loop to call our plotting function multiple times, 
isolating data points by categorical level and setting a different "marker" argument value for each one.
'''
cat_makers = [['Automatic','o'],
              ['Manual','^']]

for cat, marker in cat_makers:
    cat_df = fuel_econ_sample.loc[fuel_econ_sample['trans_type']==cat,:]
    sns.regplot(data = cat_df, x='displ', y='comb', marker = marker, fit_reg=False, x_jitter=0.04)
plt.legend(['Automatic','Manual']);
```

## Encoding via Size

```python
'''
Point size is a good encoding for numeric variables. 
Usually, we want the numeric values to be proportional to the area of the point markers; 
this is the default functionality of the "s" parameter in scatter. 
(You need to refer to "s" through a dictionary assigned to the "scatter_kws" parameter when working with regplot.)
'''

sns.regplot(data=fuel_econ_sample, x='displ',y='comb',
           x_jitter = 0.04, fit_reg = False,
           scatter_kws={'s':fuel_econ_sample['co2']/2});
#the legend is more complicated with size
#dummy reference sizes
sizes = [200,350,500]
base_color = sns.color_palette()[0]
legend_obj = []
for s in sizes:
    legend_obj.append(plt.scatter([],[],s=s/2, color = base_color))#returns scatter objects to list
plt.legend(legend_obj, sizes, title = 'CO2 (g/mi)');
```

## Encoding via color

#### Using facetgrid to encode a binary categorical variable (nominal) as color

```python
#If you have a qualitative variable, you can set different colors for different levels 
#of a categorical variable through the "hue" parameter on seaborn's FacetGrid class.

g = sns.FacetGrid(data=fuel_econ_sample, hue = 'trans_type')
g.map(sns.regplot,'displ','comb',x_jitter=0.04,fit_reg = False)
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

#### Changing the order and making further tweaks

```python
g = sns.FacetGrid(data = fuel_econ_sample, hue = 'trans_type',
                #changing order to have automatic on top/1st in legend
                 hue_order = ['Automatic', 'Manual'],
                #increasing size of chart
                 height = 4,
                #changing aspect so the chart is 50% longer than it is tall
                 aspect = 1.5)
g = g.map(sns.regplot, 'displ', 'comb', x_jitter = 0.04, fit_reg = False);
#adding legend
g.add_legend()
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

#### encoding an ordinal categorical variable with color using a sequential palette

##### ordering VClass and converting to categorical

```python
#types of sedan cars

#creating list with order from small to large
sedan_classes =['Minicompact Cars', 'Subcompact Cars', 'Compact Cars','Midsize Cars','Large Cars']

#Returns the types of sedan_classes with the categories and orderedness
#refer - https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.api.types.CategoricalDtype.html
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=sedan_classes)

# Use pandas.astype() to convert the "VClass" column from a plain object type into an ordered categorical type 
fuel_econ_sample['VClass'] = fuel_econ_sample['VClass'].astype(vclasses)
```

```python
g = sns.FacetGrid(data = fuel_econ_sample, hue = 'VClass',
                 height = 4, aspect = 1.5, 
                  #adding the palette parameter and changing hue
                  # to an ordinal variable will result in the desired
                  # sequential palette effect
                  palette = 'viridis_r')
g = g.map(sns.regplot, 'displ', 'comb', x_jitter = 0.04, fit_reg = False);
g.add_legend()
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

### encoding quantitative variables in with a sequential palette

For quantitative variables, we should not take the same approach, since FacetGrid expects any variable input for subsetting to be categorical. Instead, we can set color based on numeric value in the scatter function through the "c" parameter, much like how we set up marker sizes through "s". (Unlike with size, we don't have an easy way of setting color by numeric value through regplot due to how its "color" argument is set up.)

```python
plt.scatter(data=fuel_econ_sample, x='displ', y='comb', 
            #in scatter use c to assign the 3rd var
            #use cmap to assign the palette
            c='co2', cmap = 'viridis_r')
#add colorbar for legend
plt.colorbar(label = 'CO2 (g/mi)')
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

#### Selecting Color Palettes
If you want to change the color map for your plot, the easiest way of doing so is by using one of the built-ins from Matplotlib or Seaborn. This part of the Matplotlib documentation has a list of strings that can be understood for color mappings. For most of your purposes, stick with the palettes noted in the top few tables as built-in for Matplotlib ('viridis', etc.) or from ColorBrewer; the remaining palettes may not be as perceptually consistent. Seaborn also adds in a number of its own palettes:

- Qualitative (all up to 6 colors): 'deep', 'pastel', 'dark', 'muted', 'bright', 'colorblind'
- Sequential: 'rocket' (white-orange-red-purple-black), 'mako' (mint-green-blue-purple-black)
- Diverging: 'vlag' (blue-white-red), 'icefire' (blue-black-orange)

For all of these strings, appending '_r' reverses the palette, which is useful if a sequential or diverging palette is rendered counter to your expectations.

A color palette can be set in FacetGrid through the "palette" parameter, and in scatter through the "cmap" parameter.
