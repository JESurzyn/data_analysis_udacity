```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokemon.csv')
print(pokemon.shape)
pokemon.head()
```

## plot matrices

let's plot the six combat stats for all pokemon and their relationships to each other

```python
#using PairGrid function
variables = ['hp', 'attack', 'defense', 'speed', 'special-attack', 'special-defense']
g = sns.PairGrid(data = pokemon, vars=variables)
```

```python
#mapping charts
g.map(plt.scatter);
```

if we want to map the center charts (where the variables are shown in relation to themselves) as histograms, use the following:

```python
g=sns.PairGrid(data=pokemon, vars=variables)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
```

#### Pairgrid can be used for other types of plots

## Using Pairplot instead
PairGrid is more flexible, but if one wants scatter plots and histograms, pairplot requires fewer lines of code to write

```python
sns.pairplot(data=pokemon.loc[:,variables]);
```

```python
#cleaner to remove redundant hal
sns.pairplot(data=pokemon.loc[:,variables], corner=True);
```

## Correlation Matrices

Basic Heatmap

```python
sns.heatmap(pokemon.loc[:,variables].corr());
```

Tweaking the heatmap a bit

```python
sns.heatmap(pokemon.loc[:,variables].corr(), cmap='rocket_r', annot=True, fmt='.2f');
```

```python
sns.heatmap(pokemon.loc[:,variables].corr(), cmap='rocket_r', annot=True, fmt='.2f', vmin=0);
```
