### demonstrating distplot() and rugplot() to plot the KDE

```python
data = [0.0, 3.0, 4.5, 8.0]
plt.figure(figsize=[12,5])

#left plot: showing the kde lumps with the default settings
plt.subplot(1,3,1)
sns.distplot(data, hist=False, rug=True, rug_kws={'color': 'r'})

#central plot: kde with narrow bandwidth to show individual probability lumps
plt.subplot(1,3,2)
sns.distplot(data, hist=False, rug=True, rug_kws={'color': 'r'}, kde_kws={'bw':1})

#central plot: kde with narrow bandwidth to show individual probablility limps
plt.subplot(1,3,3)
sns.distplot(data, hist=False, rug=True, rug_kws={'color': 'r'}, kde_kws={'bw':1.5, 'kernel':'tri'})
```

```python
data = [0.0, 3.0, 4.5, 8.0]
plt.figure(figsize = [12, 5])

# left plot: showing kde lumps with the default settings
plt.subplot(1, 3, 1)
sns.kdeplot(data)
sns.rugplot(data)

# central plot: kde with narrow bandwidth to show individual probability lumps
plt.subplot(1, 3, 2)
sns.kdeplot(data, bw_adjust=1)
sns.rugplot(data)

# right plot: choosing a different, triangular kernel function (lump shape)
plt.subplot(1, 3, 3)
sns.kdeplot(data, bw_adjust=1.5, kernel='tri')
sns.rugplot(data)
```
